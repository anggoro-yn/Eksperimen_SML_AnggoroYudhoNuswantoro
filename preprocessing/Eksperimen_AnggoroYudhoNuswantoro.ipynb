# Eksperimen Sistem Machine Learning

## Data Loading

```python
import pandas as pd
df = pd.read_csv('../iris_raw/iris.csv')
```

## EDA (Eksplorasi Data)

```python
print(df.head())
print(df.info())
print(df.describe())
```

## Preprocessing

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['variety'] = le.fit_transform(df['variety'])
print(df.head()) # Sekarang kolom variety berisi angka
```

## Menyimpan Hasil

```python
import os

# Membuat folder tujuan jika belum ada
output_dir = 'iris_preprocessing'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Simpan hasil bersihnya
df.to_csv('iris_preprocessing/iris_cleaned.csv', index=False)
print("Data berhasil disimpan di folder preprocessing/iris_preprocessing/")
```
